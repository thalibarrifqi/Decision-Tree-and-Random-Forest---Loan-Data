Add the purpose column into seperate column by its categories

```python
cat_feats = ['purpose']
```

add the cat_feats column into main data frame

```python
final_data = pd.get_dummies(loans,columns=cat_feats,drop_first=True) 
```

```python
final_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_data.drop('not.fully.paid',axis=1)
y = final_data['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train, y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```

we've done with the decision tree model, lets try the random fores model and see who perform better

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rforest = RandomForestClassifier()
```

```python
rforest.fit(X_train, y_train)
```

```python
predict_forest = rforest.predict(X_test)
```

```python
print(confusion_matrix(y_test,predict_forest))
print(classification_report(y_test,predict_forest))
```

the random fores peform better
but actually it depends on what metric you are trying to optimize for
